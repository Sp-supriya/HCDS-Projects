```python
import pandas as pd

# Sample financial data
data = {
    'Quarter': ['Q1', 'Q2', 'Q3', 'Q4'] * 2,
    'Year': [2023]*4 + [2024]*4,
    'Revenue': [150000, 180000, 170000, 200000, 210000, 220000, 200000, 230000],
    'Expenses': [90000, 95000, 100000, 110000, 105000, 108000, 102000, 115000],
    'Profit': [60000, 85000, 70000, 90000, 105000, 112000, 98000, 115000]
}

df = pd.DataFrame(data)
df.to_csv("financial_data.csv", index=False)

from google.colab import files
files.download("financial_data.csv")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Upload the CSV
from google.colab import files
uploaded = files.upload()

# Load data
df = pd.read_csv("financial_data.csv")

# Display head
print("ðŸ“Š Dataset Preview:")
display(df.head())

# Summary stats
print("\nðŸ“ˆ Summary Statistics:")
display(df.describe())

# Profit margin calculation
df['Profit_Margin'] = (df['Profit'] / df['Revenue']) * 100
```

```python
# Revenue, Expenses, Profit over time
fig = px.line(df, x="Quarter", y=["Revenue", "Expenses", "Profit"], color='Year', title="Quarterly Financial Overview")
fig.show()

# Profit Margin
sns.barplot(data=df, x="Quarter", y="Profit_Margin", hue="Year")
plt.title("Profit Margin by Quarter")
plt.ylabel("Profit Margin (%)")
plt.show()
```
