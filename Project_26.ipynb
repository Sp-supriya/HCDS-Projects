```python
import pandas as pd

# Create sample drug data
data = {
    "drug_name": ["DrugA", "DrugB", "DrugC", "DrugA", "DrugB", "DrugC", "DrugA", "DrugB"],
    "age_group": ["Adult", "Senior", "Adult", "Teen", "Senior", "Teen", "Adult", "Teen"],
    "side_effect": ["Nausea", "Headache", "Dizziness", "Nausea", "Nausea", "Headache", "Dizziness", "Headache"]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Save CSV
df.to_csv("drug_side_effects.csv", index=False)

# Load it back
df = pd.read_csv("drug_side_effects.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical data
le_drug = LabelEncoder()
le_age = LabelEncoder()
le_effect = LabelEncoder()

df["drug_encoded"] = le_drug.fit_transform(df["drug_name"])
df["age_encoded"] = le_age.fit_transform(df["age_group"])
df["effect_encoded"] = le_effect.fit_transform(df["side_effect"])

# Features and label
X = df[["drug_encoded", "age_encoded"]]
y = df["effect_encoded"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

# Predict
y_pred = clf.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=le_effect.classes_))
```

```python
# Add predictions to DataFrame
df_test = X_test.copy()
df_test["true_label"] = le_effect.inverse_transform(y_test)
df_test["predicted_label"] = le_effect.inverse_transform(y_pred)

# Save as CSV
df_test.to_csv("side_effect_predictions.csv", index=False)

# Download file
from google.colab import files
files.download("side_effect_predictions.csv")
```
