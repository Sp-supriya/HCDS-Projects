```python
import pandas as pd

# Sample feedback
data = {
    'employee_feedback': [
        "I enjoy the hybrid work model, but we need better virtual meeting tools.",
        "Flexible seating helps collaboration, but sometimes it's hard to find a spot.",
        "Remote work is productive but we miss team bonding events.",
        "Better communication tools would enhance flexible work management."
    ]
}

df = pd.DataFrame(data)
df.to_csv('office_feedback.csv', index=False)
print("CSV created.")

from google.colab import files
files.download('office_feedback.csv')
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Load feedback
df = pd.read_csv('office_feedback.csv')

# Text Vectorization
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['employee_feedback'])

# Topic Modeling
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(X)

# Display Topics
for idx, topic in enumerate(lda.components_):
    print(f"\nTopic #{idx + 1}:")
    print([vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-5:]])
```

```python
import pandas as pd

# Sample feedback
data = {
    "employee_feedback": [
        "The open floor plan is great for collaboration.",
        "I find it hard to concentrate with too much noise.",
        "Flexibility to work from home is a huge plus.",
        "More private meeting rooms are needed.",
        "I enjoy the communal kitchen and coffee area."
    ]
}

# Save to CSV
df = pd.DataFrame(data)
df.to_csv("office_feedback.csv", index=False)

# Load it
df = pd.read_csv("office_feedback.csv")
df.head()
```

```python
from transformers import pipeline

# Load summarizer
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Join all feedback into one paragraph
full_text = " ".join(df['employee_feedback'].tolist())

# Summarize
summary = summarizer(full_text, max_length=60, min_length=20, do_sample=False)
print("Summary:\n", summary[0]['summary_text'])
```

```python
with open("summary_flexible_office.txt", "w") as f:
    f.write(summary[0]['summary_text'])

from google.colab import files
files.download("summary_flexible_office.txt")
```
