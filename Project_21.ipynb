```python
from transformers import pipeline
import pandas as pd
```

```python
# Load zero-shot classification pipeline for emotion analysis
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)
```

```python
def analyze_emotions(text):
    results = classifier(text)[0]
    sorted_results = sorted(results, key=lambda x: x['score'], reverse=True)
    return sorted_results[0]['label'], sorted_results[0]['score']
```

```python
comments = [
    "I'm really happy with the service!",
    "This is so frustrating and disappointing.",
    "I'm feeling a bit anxious about the results.",
    "Absolutely loved the product!",
    "I feel ignored and unimportant."
]

for comment in comments:
    emotion, confidence = analyze_emotions(comment)
    print(f"Comment: {comment}\nPredicted Emotion: {emotion} (Confidence: {confidence:.2f})\n")
```
