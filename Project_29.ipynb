```python
import pandas as pd

# Sample data creation
data = {
    'Campaign_ID': [1, 2, 3, 4, 5],
    'Campaign_Name': ['Campaign A', 'Campaign B', 'Campaign C', 'Campaign D', 'Campaign E'],
    'Channel': ['Social Media', 'Email', 'Search Ads', 'Social Media', 'Email'],
    'Spend': [5000, 3000, 7000, 4000, 2000],
    'Impressions': [200000, 150000, 250000, 220000, 180000],
    'Clicks': [10000, 5000, 15000, 11000, 7000],
    'Conversions': [300, 100, 400, 350, 150],
    'Revenue': [9000, 3500, 12000, 10500, 4500],
    'Date': ['2023-01-01', '2023-01-05', '2023-02-01', '2023-02-10', '2023-03-01']
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame as CSV
df.to_csv('/content/marketing_campaign_data.csv', index=False)
```

```python
# Load the CSV file
df = pd.read_csv('/content/marketing_campaign_data.csv')

# Show the first few rows
df.head()
```

```python
# Calculate additional metrics
df['CTR'] = df['Clicks'] / df['Impressions'] * 100  # Click-Through Rate (CTR)
df['Conversion_Rate'] = df['Conversions'] / df['Clicks'] * 100  # Conversion Rate
df['CPC'] = df['Spend'] / df['Clicks']  # Cost per Click (CPC)
df['ROI'] = (df['Revenue'] - df['Spend']) / df['Spend'] * 100  # Return on Investment (ROI)

# Preview the updated DataFrame
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='Campaign_Name', y='ROI', data=df, palette='Blues_d')
plt.title('Return on Investment (ROI) by Campaign')
plt.xlabel('Campaign')
plt.ylabel('ROI (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set up the visualization
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='CTR', data=df, marker='o', color='green')
plt.title('Click-Through Rate (CTR) Over Time')
plt.xlabel('Date')
plt.ylabel('CTR (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Pie Chart for Channel Distribution
channel_dist = df['Channel'].value_counts()
plt.figure(figsize=(8, 8))
channel_dist.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3', len(channel_dist)))
plt.title('Campaign Channel Distribution')
plt.ylabel('')
plt.show()
```

```python
# Save the updated dataframe with metrics to CSV
df.to_csv('/content/marketing_campaign_data_with_metrics.csv', index=False)
```
