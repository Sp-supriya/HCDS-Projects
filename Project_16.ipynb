```python
import pandas as pd
from dash import Dash, dcc, html, Input, Output
import plotly.express as px
from flask_ngrok import run_with_ngrok

# Load dataset
df = pd.read_csv('retail_data.csv')

# Initialize Dash app
app = Dash(__name__)
run_with_ngrok(app)

app.layout = html.Div([
    html.H1("Retail Analytics Dashboard", style={'textAlign': 'center'}),

    html.Label("Select Category:"),
    dcc.Dropdown(
        id='category-dropdown',
        options=[{'label': cat, 'value': cat} for cat in df['Category'].unique()],
        value=df['Category'].unique()[0]
    ),

    dcc.Graph(id='revenue-graph'),

    dcc.Graph(id='units-sold-graph')
])

@app.callback(
    [Output('revenue-graph', 'figure'),
     Output('units-sold-graph', 'figure')],
    [Input('category-dropdown', 'value')]
)
def update_graphs(selected_category):
    filtered_df = df[df['Category'] == selected_category]

    revenue_fig = px.bar(filtered_df, x='Date', y='Revenue', color='Product', title='Revenue Over Time')
    units_fig = px.line(filtered_df, x='Date', y='Units_Sold', color='Product', title='Units Sold Over Time')

    return revenue_fig, units_fig

app.run()
```

```python
from dash import Dash, dcc, html
import plotly.express as px
import pandas as pd

# Expanded sample data for the dashboard
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05', '2024-01-06', '2024-01-07',
             '2024-01-08', '2024-01-09', '2024-01-10'],
    'Product': ['Shampoo', 'Soap', 'Notebook', 'Toothpaste', 'Pen', 'Charger', 'Headphones', 'Socks',
                'T-Shirt', 'Mug'],
    'Category': ['Personal Care', 'Personal Care', 'Stationery', 'Personal Care', 'Stationery', 'Electronics',
                 'Electronics', 'Apparel', 'Apparel', 'Home Goods'],
    'Revenue': [1500, 800, 1200, 950, 400, 500, 1000, 150, 600, 300],
    'Units_Sold': [30, 40, 20, 25, 50, 10, 15, 100, 70, 25],
    'Region': ['South', 'North', 'West', 'East', 'South', 'North', 'East', 'West', 'South', 'East']
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Create the Dash app
app = Dash(__name__)

# Create figures for the dashboard
revenue_fig = px.bar(df, x='Product', y='Revenue', title="Product Revenue")
units_fig = px.bar(df, x='Product', y='Units_Sold', title="Units Sold")

# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("Retail Analytics Dashboard"),
    dcc.Graph(figure=revenue_fig),
    dcc.Graph(figure=units_fig)
])

# Run the server
app.run()
```
