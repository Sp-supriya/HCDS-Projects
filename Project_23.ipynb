# Automate detection of different sentiments from paragraphs and predict overall sentiment

```python
from transformers import pipeline
import pandas as pd

# Load the pre-trained sentiment analysis model
sentiment_analyzer = pipeline("sentiment-analysis")
```

```python
# Sample paragraphs (could be loaded from a file or another source)
paragraphs = [
    "I absolutely love the new design of the product. It's amazing!",
    "The product is not bad, but it could use some improvements.",
    "I am really disappointed with the product. It didn't meet my expectations.",
    "The customer service was very helpful and resolved my issue quickly."
]
```

```python
# Analyze sentiment for each paragraph
sentiment_results = []

for paragraph in paragraphs:
    result = sentiment_analyzer(paragraph)[0]  # Get the sentiment prediction
    sentiment_results.append({
        'Paragraph': paragraph,
        'Sentiment': result['label'],
        'Confidence': result['score']
    })

# Create a DataFrame to hold the results
df = pd.DataFrame(sentiment_results)
```

```python
# Count sentiment labels and predict overall sentiment
sentiment_counts = df['Sentiment'].value_counts()

# Predict overall sentiment
overall_sentiment = sentiment_counts.idxmax()  # Get the most frequent sentiment
overall_confidence = sentiment_counts.max() / len(df)  # Proportion of most frequent sentiment

# Add overall sentiment information to the DataFrame
df['Overall_Sentiment'] = overall_sentiment
df['Overall_Confidence'] = overall_confidence
```
