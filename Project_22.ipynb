```python
# Step 1: Install Java 17 and dependencies
!sudo apt-get update
!sudo apt-get install openjdk-17-jdk -y
import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-17-openjdk-amd64"
os.environ["PATH"] = os.environ["JAVA_HOME"] + "/bin:" + os.environ["PATH"]

# Step 2: Install libraries
!pip install -q language-tool-python transformers pandas

# Step 3: Import libraries
import language_tool_python
import pandas as pd
from transformers import pipeline

# Step 4: Initialize tools
tool = language_tool_python.LanguageTool('en-US')
semantic_check = pipeline("text-classification", model="textattack/roberta-base-RTE")

# Step 5: Sample input
sentences = [
    "He go to the school everyday.",
    "Their coming to the party tomorrow.",
    "She has finish her homework."
]

# Step 6: Check and correct
results = []
for sentence in sentences:
    matches = tool.check(sentence)
    corrected = language_tool_python.utils.correct(sentence, matches)
    semantic_result = semantic_check(sentence)[0]
    results.append({
        "Original": sentence,
        "Corrected": corrected,
        "Semantic_Label": semantic_result['label'],
        "Confidence": semantic_result['score']
    })

# Step 7: Save to CSV
df = pd.DataFrame(results)
df.to_csv("semantic_proofreading_results.csv", index=False)

# Step 8: Download CSV
from google.colab import files
files.download("semantic_proofreading_results.csv")
```

```python
import language_tool_python
from transformers import pipeline

# Initialize tools
tool = language_tool_python.LanguageTool('en-US')
semantic_check = pipeline("text-classification", model="textattack/roberta-base-RTE")

# Load your sample sentences
with open("sample_sentences.txt", "r") as file:
    sentences = file.readlines()

# Step 1: Grammar + Semantic Analysis
for i, sentence in enumerate(sentences, 1):
    sentence = sentence.strip()
    if not sentence:
        continue

    # Grammar errors
    grammar_issues = tool.check(sentence)
    print(f"Sentence {i}: {sentence}")

    if grammar_issues:
        print("⛔ Grammar Issues:")
        for issue in grammar_issues:
            print(f" - {issue.message}")
    else:
        print("✅ No grammar issues.")

    # Semantic analysis - check if sentence is plausible (entailment vs contradiction)
    result = semantic_check(f"The sentence: '{sentence}' makes sense.")
    print(f"🤖 Semantic Check: {result[0]['label']} (score: {result[0]['score']:.2f})\n")
```

```python
import pandas as pd

results = []

for sentence in sentences:
    sentence = sentence.strip()
    if not sentence:
        continue
    issues = tool.check(sentence)
    issue_msgs = [i.message for i in issues]
    semantic = semantic_check(f"The sentence: '{sentence}' makes sense.")
    results.append({
        "Sentence": sentence,
        "Grammar Issues": "; ".join(issue_msgs) if issue_msgs else "None",
        "Semantic Check": semantic[0]['label'],
        "Confidence": round(semantic[0]['score'], 2)
    })

df = pd.DataFrame(results)
df.to_csv("semantic_proofreading_report.csv", index=False)

# To download:
from google.colab import files
files.download("semantic_proofreading_report.csv")
```
